```python
import os
import torch
import numpy as np
from torch import nn, optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix
import timm
from PIL import Image
import tensorflow_datasets as tfds
import logging
import random
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split # Imported twice, redundant
import glob # For listing files
from typing import Optional, Tuple
from torch.nn.modules.transformer import TransformerEncoderLayer
import torch.nn.functional as F


# Configuration class
# Configuration class
class Config:
    IMG_SIZE = 224  # Keep IMG_SIZE as 224, EuroSAT images are 64x64, we will upscale them.
    BATCH_SIZE = 32
    NUM_CLASSES = 10 # EuroSAT has 10 classes
    EPOCHS_PHASE1 = 5
    EPOCHS_PHASE2 = 10
    PREPROCESSED_DIR = "/content/drive/MyDrive/preprocessed_eurosat" # Modified PREPROCESSED_DIR for eurosat
    DATASET_FRACTION = 1.0
    LEARNING_RATE_PHASE1 = 1e-3
    LEARNING_RATE_PHASE2 = 1e-4
    EARLY_STOPPING_PATIENCE = 3
    CLIP_GRAD_NORM = 1.0
    SEED = 42
    TRAIN_SIZE = 0.7
    VAL_SIZE = 0.15
    TEST_SIZE = 0.15
    MEAN = [0.485, 0.456, 0.406]  # ImageNet mean
    STD = [0.229, 0.224, 0.225]  # ImageNet std

config = Config()



# DATASET_DIR and PREPROCESSED_DIR are now less relevant as we use TFDS, but we keep PREPROCESSED_DIR for potential future use.
PREPROCESSED_DIR = config.PREPROCESSED_DIR # Modified PREPROCESSED_DIR for eurosat


# Setup logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Set random seeds for reproducibility
def set_seed(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(config.SEED)

# Device configuration
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
logging.info(f"Using device: {DEVICE}")


from google.colab import drive
drive.mount('/content/drive')

# Constants (modified for EuroSAT) - mostly in config now


# Dataset class (Modified name to EuroSATDataset and adapted for EuroSAT data)
class EuroSATDataset(Dataset):
    def __init__(self, images, labels, transforms):
        self.images = images
        self.labels = labels
        self.transforms = transforms

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = self.images[idx]
        image = Image.fromarray(image)
        image = self.transforms(image)
        label = torch.tensor(self.labels[idx], dtype=torch.long)
        return image, label





from sklearn.model_selection import train_test_split

def load_or_preprocess_data(dataset_fraction=1.0, train_split=0.8, val_split=0.1, test_split=0.1):
    """Load or preprocess the EuroSAT dataset from TFDS."""
    global NUM_CLASSES
    if train_split + val_split + test_split != 1.0:
        raise ValueError("Train, validation, and test splits must sum to 1.0")

    try:
        if os.path.exists(os.path.join(PREPROCESSED_DIR, "train_data.npz")): # Check for preprocessed EuroSAT data
            print("Loading preprocessed EuroSAT data from .npz files...")
            with np.load(os.path.join(PREPROCESSED_DIR, "train_data.npz")) as data:
                train_images, train_labels = data['images'], data['labels']
            with np.load(os.path.join(PREPROCESSED_DIR, "val_data.npz")) as data:
                val_images, val_labels = data['images'], data['labels']
            with np.load(os.path.join(PREPROCESSED_DIR, "test_data.npz")) as data:
                test_images, test_labels = data['images'], data['labels']
            NUM_CLASSES = len(np.unique(train_labels))
        else:
            print("Processing EuroSAT dataset from TFDS...")
            # Load EuroSAT dataset from TFDS
            ds_builder = tfds.builder('eurosat/rgb')
            ds_builder.download_and_prepare()
            eurosat_ds = ds_builder.as_dataset(shuffle_files=True)

            # Extract train dataset (EuroSAT TFDS only has 'train' split)
            full_train_ds = eurosat_ds['train'] # Load the full 'train' split

            # Convert TFDS dataset to lists of images and labels
            def tfds_to_numpy(tfds_dataset):
                images = []
                labels = []
                for example in tqdm(tfds_dataset, desc="Converting TFDS to NumPy"):
                    image = example['image'].numpy()
                    label = example['label'].numpy()
                    images.append(image)
                    labels.append(label)
                return images, labels

            full_train_images_list, full_train_labels_list = tfds_to_numpy(full_train_ds) # Process the full train dataset

            full_train_images = np.array(full_train_images_list)
            full_train_labels = np.array(full_train_labels_list)


            NUM_CLASSES = len(np.unique(full_train_labels))

            # Split the full training set into train, validation, and test sets
            train_images, test_images, train_labels, test_labels = train_test_split(
                full_train_images, full_train_labels, test_size=test_split, stratify=full_train_labels, random_state=config.SEED
            )
            train_images, val_images, train_labels, val_labels = train_test_split(
                train_images, train_labels, test_size=val_split / (train_split + val_split), stratify=train_labels, random_state=config.SEED
            )

            os.makedirs(PREPROCESSED_DIR, exist_ok=True) # Ensure directory exists
            np.savez_compressed(os.path.join(PREPROCESSED_DIR, "train_data.npz"), images=train_images, labels=train_labels)
            np.savez_compressed(os.path.join(PREPROCESSED_DIR, "val_data.npz"), images=val_images, labels=val_labels)
            np.savez_compressed(os.path.join(PREPROCESSED_DIR, "test_data.npz"), images=test_images, labels=test_labels)


    except Exception as e:
        print(f"Error loading or preprocessing EuroSAT data: {e}")
        raise e

    return train_images, train_labels, val_images, val_labels, test_images, test_labels

# Data transformations
# Data transformations
train_transforms = transforms.Compose([
    transforms.RandomResizedCrop(config.IMG_SIZE, scale=(0.8, 1.0)),
    transforms.RandomRotation(degrees=10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.5),
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1), scale=(0.9, 1.1), shear=10),
    transforms.ToTensor(),
    transforms.Normalize(mean=config.MEAN, std=config.STD)
])


val_transforms = transforms.Compose([
    transforms.Resize((config.IMG_SIZE, config.IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=config.MEAN, std=config.STD)
])

# CBAM Module (Channel and Spatial Attention Module)
class ChannelAttention(nn.Module):
    def __init__(self, in_planes, ratio=16):
        super(ChannelAttention, self).__init__()
        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.max_pool = nn.AdaptiveMaxPool2d(1)

        self.fc1   = nn.Conv2d(in_planes, in_planes // ratio, 1, bias=False)
        self.relu1 = nn.ReLU()
        self.fc2   = nn.Conv2d(in_planes // ratio, in_planes, 1, bias=False)

        self.sigmoid = nn.Sigmoid()

        # Initialize weights
        nn.init.xavier_uniform_(self.fc1.weight)
        nn.init.xavier_uniform_(self.fc2.weight)

    def forward(self, x):
        avg_out = self.fc2(self.relu1(self.fc1(self.avg_pool(x))))
        max_out = self.fc2(self.relu1(self.fc1(self.max_pool(x))))
        out = avg_out + max_out
        return self.sigmoid(out)

class SpatialAttention(nn.Module):
    def __init__(self, kernel_size=7):
        super(SpatialAttention, self).__init__()

        assert kernel_size in (3, 7), 'kernel size must be 3 or 7'
        padding = 3 if kernel_size == 7 else 1

        self.conv1 = nn.Conv2d(2, 1, kernel_size, padding=padding, bias=False)
        self.sigmoid = nn.Sigmoid()

        # Initialize weights
        nn.init.xavier_uniform_(self.conv1.weight)

    def forward(self, x):
        avg_out = torch.mean(x, dim=1, keepdim=True)
        max_out, _ = torch.max(x, dim=1, keepdim=True)
        x = torch.cat([avg_out, max_out], dim=1)
        x = self.conv1(x)
        return self.sigmoid(x)

class CBAM(nn.Module):
    """Convolutional Block Attention Module (CBAM)."""
    def __init__(self, channels, reduction_ratio=16, kernel_size=7):
        super(CBAM, self).__init__()
        self.channelattention = ChannelAttention(channels, ratio=reduction_ratio)
        self.spatialattention = SpatialAttention(kernel_size=kernel_size)
        self.dropout = nn.Dropout(0.3)
        self._initialize_weights()

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)

    def forward(self, x):
        x = x * self.channelattention(x)
        x = x * self.spatialattention(x)
        return self.dropout(x)

# Positional Encoding
class PositionalEncoding(nn.Module):
    """Positional encoding."""
    def __init__(self, d_model, dropout=0.1, max_len=5000):
        super(PositionalEncoding, self).__init__()
        self.dropout = nn.Dropout(p=dropout)

        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-np.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0).transpose(0, 1)
        self.register_buffer('pe', pe)

    def forward(self, x):
        x = x + self.pe[:x.size(0), :]
        return self.dropout(x)

from torch.nn.modules.transformer import TransformerEncoderLayer
import torch.nn.functional as F
from typing import Optional, Tuple

from torch.nn.modules.transformer import TransformerEncoderLayer
import torch.nn.functional as F
from typing import Optional, Tuple

class TransformerEncoderLayerWithAttention(TransformerEncoderLayer):
    def __init__(self, d_model, nhead, dim_feedforward=2048, dropout=0.1, activation=F.relu,
                 layer_norm_eps=1e-5, batch_first=False, norm_first=False,
                 device=None, dtype=None) -> None:
        factory_kwargs = {'device': device, 'dtype': dtype}
        super().__init__(d_model=d_model, nhead=nhead, dim_feedforward=dim_feedforward, dropout=dropout, activation=activation,
                         layer_norm_eps=layer_norm_eps, batch_first=batch_first, norm_first=norm_first,
                         **factory_kwargs)

    def forward(self, src: torch.Tensor, src_mask: Optional[torch.Tensor] = None, src_key_padding_mask: Optional[torch.Tensor] = None) -> Tuple[torch.Tensor, torch.Tensor]:
        """Pass inputs through the encoder layer.

        Args:
            src: the sequence to the encoder layer (required).
            src_mask: the mask for the src sequence (optional).
            src_key_padding_mask: the mask for the src keys per batch (optional).

        Shape:
            see the docs in TransformerEncoderLayer class.

        Returns:
            output torch.Tensor: The output of the encoder layer.
            attn_weights torch.Tensor: The attention weights of the MHA module.
        """
        x = src
        attn_output, attn_weights = self.self_attn(x, x, x,
                                          attn_mask=src_mask,
                                          key_padding_mask=src_key_padding_mask,
                                          need_weights=True)  # need_weights=True to return attention weights
        x = x + self.dropout1(attn_output)
        x = self.norm1(x)
        x2 = self.linear2(self.dropout(self.activation(self.linear1(x))))  # Use self.activation instead of self.relu
        x = x + self.dropout2(x2)
        x = self.norm2(x)
        return x, attn_weights  # Return attention weights along with the output

# Late Fusion Model with Transformer
class LateFusionWithTransformer(nn.Module):
    def __init__(self, num_classes, embed_dim=256, num_heads=4, num_layers=2):
        super(LateFusionWithTransformer, self).__init__()
        self.device = DEVICE
        self.swin = timm.create_model('swin_tiny_patch4_window7_224', pretrained=True, num_classes=num_classes)
        self.efficientnet = timm.create_model('efficientnet_b0', pretrained=True, num_classes=num_classes)
        self.cbam = CBAM(channels=1280)
        self.swin.to(self.device)
        self.efficientnet.to(self.device)
        self.cbam.to(self.device)

        # Projection layers with Xavier init
        self.swin_proj = nn.Linear(num_classes, embed_dim)
        self.swin_dropout = nn.Dropout(0.3)
        self.eff_proj = nn.Linear(num_classes, embed_dim)
        self.eff_dropout = nn.Dropout(0.3)

        # Transformer Encoder avec notre couche modifiée
        encoder_layer = TransformerEncoderLayerWithAttention( # Utilisation de TransformerEncoderLayerWithAttention
            d_model=embed_dim,
            nhead=num_heads,
            batch_first=True,
            dropout=0.2
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)

        # Classifier with Xavier init
        self.classifier = nn.Linear(embed_dim, num_classes)
        self.classifier_dropout = nn.Dropout(0.3)

        self.positional_encoding = PositionalEncoding(embed_dim)

        self.to(self.device)

        # Initialize projection and classifier layers with Xavier uniform
        nn.init.xavier_uniform_(self.swin_proj.weight)
        nn.init.xavier_uniform_(self.eff_proj.weight)
        nn.init.xavier_uniform_(self.classifier.weight)

    def forward(self, x):
        # Swin Transformer
        swin_out = self.swin(x)
        swin_proj = self.swin_proj(swin_out)
        swin_proj = self.swin_dropout(swin_proj)

        # EfficientNet with CBAM
        eff_features = self.efficientnet.forward_features(x)
        eff_features = self.cbam(eff_features)
        eff_out = self.efficientnet.classifier(nn.functional.adaptive_avg_pool2d(eff_features, (1, 1)).flatten(1))
        eff_proj = self.eff_proj(eff_out)
        eff_proj = self.eff_dropout(eff_proj)

        # Combine features
        combined = torch.stack([swin_proj, eff_proj], dim=1)
        combined = self.positional_encoding(combined)

        # Transformer Encoder
        attn_weights_list = [] # Liste pour stocker les poids d'attention de chaque couche
        fused_features = combined
        for layer in self.transformer_encoder.layers: # Iterer explicitement sur les couches pour récupérer les poids
            fused_features, attn_weights = layer(fused_features) # On récupère les poids ici
            attn_weights_list.append(attn_weights)


        # Aggregate features
        fused_features = torch.mean(fused_features, dim=1)
        fused_features = self.classifier_dropout(fused_features)

        # Classifier
        output = self.classifier(fused_features)

        return output, None, None, attn_weights_list # Retourne la liste des poids d'attention


    def get_logits(self, x):
        """Returns logits from Swin and EfficientNet before fusion."""
        with torch.no_grad():
            swin_out = self.swin(x)
            eff_features = self.efficientnet.forward_features(x)
            eff_features = self.cbam(eff_features)
            eff_out = self.efficientnet.classifier(nn.functional.adaptive_avg_pool2d(eff_features, (1, 1)).flatten(1))
        return swin_out, eff_out

def plot_attention_weights(model, dataloader, device, num_samples=100):
    """Plots the attention weights of the Transformer layers."""
    model.eval()
    attention_weights_list = []
    labels_list = []

    with torch.no_grad():
        for images, labels in dataloader:
            images = images.to(device)
            _, _, _, attn_weights_batch = model(images) # Récupérer les poids d'attention
            attention_weights_list.append(attn_weights_batch)
            labels_list.append(labels.cpu().numpy())
            if len(attention_weights_list) * dataloader.batch_size >= num_samples:
                break

    # Concaténer les poids d'attention et les labels
    attention_weights_all = [torch.cat([attn[layer_idx] for attn in attention_weights_list], dim=0).cpu().numpy() for layer_idx in range(len(attention_weights_list[0]))] # Poids par couche
    labels_np = np.concatenate(labels_list, axis=0)[:num_samples]

    num_layers = len(attention_weights_all)

    for layer_idx in range(num_layers):
        avg_attn_weights = np.mean(attention_weights_all[layer_idx][:num_samples], axis=0) # Moyenne des poids pour cette couche

        plt.figure(figsize=(8, 6))
        sns.heatmap(avg_attn_weights, annot=True, cmap="viridis", fmt=".2f",
                    xticklabels=['Swin', 'EfficientNet'],
                    yticklabels=['Swin', 'EfficientNet'])
        plt.xlabel("Source (Clés/Valeurs)")
        plt.ylabel("Requête")
        plt.title(f"Layer {layer_idx+1} - Average Attention Weights - EuroSAT") # Titre plus informatif
        plt.savefig(f'attention_layer_{layer_idx+1}_eurosat.png')
        plt.show()


from torch.cuda.amp import GradScaler, autocast

scaler = GradScaler()

def train_one_epoch(model, dataloader, criterion, optimizer, device, scheduler, clip_value):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for inputs, labels in tqdm(dataloader, desc="Training"):
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()

        with autocast():
            outputs, _, _, _ = model(inputs)
            loss = criterion(outputs, labels)

        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        nn.utils.clip_grad_norm_(model.parameters(), clip_value)
        scaler.step(optimizer)
        scaler.update()

        scheduler.step()

        running_loss += loss.item() * inputs.size(0)
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    epoch_loss = running_loss / total
    epoch_acc = correct / total
    return epoch_loss, epoch_acc


def validate(model, dataloader, criterion, device):
    """Validates the model."""
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():
        for inputs, labels in tqdm(dataloader, desc="Validation"):
            inputs, labels = inputs.to(device), labels.to(device)
            outputs, _, _, _ = model(inputs) # Modified to get only output from forward
            loss = criterion(outputs, labels)

            running_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    epoch_loss = running_loss / total
    epoch_acc = correct / total
    return epoch_loss, epoch_acc


def train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs, device, patience, clip_value, model_name="best_model_late_fusion_eurosat.pth"):
    """
    Trains the model with early stopping and tracks loss/accuracy.
    """
    best_val_acc = 0.0
    patience_counter = 0
    train_losses = []
    val_losses = []
    train_accs = []
    val_accs = []

    for epoch in range(num_epochs):
        logging.info(f"Epoch {epoch + 1}/{num_epochs}")

        train_loss, train_acc = train_one_epoch(model, train_loader, criterion, optimizer, device, scheduler, clip_value)
        logging.info(f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}")

        val_loss, val_acc = validate(model, val_loader, criterion, device)
        logging.info(f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}")

        print(f"Epoch {epoch + 1}/{num_epochs}, Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}")

        train_losses.append(train_loss)
        val_losses.append(val_loss)
        train_accs.append(train_acc)
        val_accs.append(val_acc)

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            patience_counter = 0
            torch.save(model.state_dict(), model_name)
            logging.info("Best model saved!")
        else:
            patience_counter += 1
            logging.info(f"Early stopping patience counter: {patience_counter}/{patience}")
            if patience_counter >= patience:
                logging.info("Early stopping triggered.")
                break
        torch.cuda.empty_cache() # Clear GPU memory after each epoch

    logging.info("Loading best model for evaluation")
    model.load_state_dict(torch.load(model_name))
    model.eval()
    return model, train_losses, val_losses, train_accs, val_accs


def evaluate_model(model, test_loader, train_loader, val_loader, device, num_classes):
    """Evaluates the model on the test set and displays metrics."""
    model.eval()

    # Evaluate on val and test sets
    for loader, name in zip([val_loader, test_loader], ["Validation", "Test"]): # Removed train_loader from evaluation sets.
        all_preds = []
        all_labels = []

        with torch.no_grad():
            for images, labels in tqdm(loader, desc=f"Testing on {name}"):
                images, labels = images.to(device), labels.to(device)
                outputs, _, _, _ = model(images) # Modified to get only output from forward
                preds = outputs.argmax(1)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

        # Classification Report
        logging.info(f"{name} Classification Report:")
        report = classification_report(all_labels, all_preds, target_names=[str(i) for i in range(num_classes)], digits=4)
        print(report)

        # Confusion Matrix (Percentage)
        cm = confusion_matrix(all_labels, all_preds, normalize='true')
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt=".2%", cmap="Blues")
        plt.xlabel("Predicted Labels")
        plt.ylabel("True Labels")
        plt.title(f"{name} Confusion Matrix (Percentage) - EuroSAT") # Added EuroSAT to title
        plt.show()

        # Save the Confusion matrix
        plt.savefig(f'{name.lower()}_eurosat_confusion_matrix.png') # Added eurosat to filename

        # Save the Classification report
        with open(f"{name.lower()}_eurosat_classification_report.txt", "w") as file: # Added eurosat to filename
            file.write(report)


def plot_loss_curves(train_losses, val_losses, phase):
    """Plots training and validation loss curves."""
    plt.figure(figsize=(10, 6))
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(f'Loss Curves (Phase {phase}) - EuroSAT') # Added EuroSAT to title
    plt.legend()
    plt.savefig(f'loss_curves_phase_{phase}_eurosat.png') # Added eurosat to filename
    plt.show()

def plot_accuracy_curves(train_accs, val_accs, phase):
    """Plots training and validation accuracy curves."""
    plt.figure(figsize=(10, 6))
    plt.plot(train_accs, label='Train Accuracy')
    plt.plot(val_accs, label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title(f'Accuracy Curves (Phase {phase}) - EuroSAT') # Added EuroSAT to title
    plt.legend()
    plt.savefig(f'accuracy_curves_phase_{phase}_eurosat.png') # Added eurosat to filename
    plt.show()


if __name__ == "__main__":

    # Do not delete preprocessed files if directory exists
    # import shutil
    # if os.path.exists(PREPROCESSED_DIR):
    #     shutil.rmtree(PREPROCESSED_DIR)
    #     print(f"Deleted existing preprocessed directory: {PREPROCESSED_DIR}")

    # Charger les données prétraitées EuroSAT
    train_images, train_labels, val_images, val_labels, test_images, test_labels = load_or_preprocess_data(
        dataset_fraction=config.DATASET_FRACTION,
        train_split=config.TRAIN_SIZE,
        val_split=config.VAL_SIZE,
        test_split=config.TEST_SIZE
    )

    train_dataset = EuroSATDataset(train_images, train_labels, train_transforms) # Use EuroSATDataset
    val_dataset = EuroSATDataset(val_images, val_labels, val_transforms) # Use EuroSATDataset
    test_dataset = EuroSATDataset(test_images, test_labels, val_transforms) # Use EuroSATDataset

    train_loader = DataLoader(train_dataset, batch_size=config.BATCH_SIZE, shuffle=True, pin_memory=True, num_workers=2)
    val_loader = DataLoader(val_dataset, batch_size=config.BATCH_SIZE, shuffle=False, pin_memory=True, num_workers=2)
    test_loader = DataLoader(test_dataset, batch_size=config.BATCH_SIZE, shuffle=False, pin_memory=True, num_workers=2)

    # Initialiser le modèle
    model = LateFusionWithTransformer(config.NUM_CLASSES).to(DEVICE)

    criterion = nn.CrossEntropyLoss()

    # --- Phase 1 : Entraînement du module de fusion ---
    logging.info("Phase 1: Entraînement du module de fusion")

    # Freeze Swin and EfficientNet parameters
    for param in model.swin.parameters():
        param.requires_grad = False
    for param in model.efficientnet.parameters():
        param.requires_grad = False

    optimizer = optim.Adam(
        [
            {'params': model.cbam.parameters()},
            {'params': model.swin_proj.parameters()},
            {'params': model.eff_proj.parameters()},
            {'params': model.transformer_encoder.parameters()},
            {'params': model.classifier.parameters()}
        ],
        lr=config.LEARNING_RATE_PHASE1
    )
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=config.EPOCHS_PHASE1)

    model, train_losses_phase1, val_losses_phase1, train_accs_phase1, val_accs_phase1 = train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, config.EPOCHS_PHASE1, DEVICE, config.EARLY_STOPPING_PATIENCE, config.CLIP_GRAD_NORM, model_name="best_model_late_fusion_eurosat_phase1.pth") # Modified model name

    # Plot loss and accuracy curves for phase 1
    #plot_loss_curves(train_losses_phase1, val_losses_phase1, 1)
    #plot_accuracy_curves(train_accs_phase1, val_accs_phase1, 1)

    # --- Phase 2 : Fine-tuning de l'ensemble du modèle ---
    logging.info("\nPhase 2: Fine-tuning de l'ensemble du modèle")

    # Unfreeze all parameters
    for param in model.parameters():
        param.requires_grad = True

    optimizer = optim.Adam(model.parameters(), lr=config.LEARNING_RATE_PHASE2)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=config.EPOCHS_PHASE2)

    model, train_losses_phase2, val_losses_phase2, train_accs_phase2, val_accs_phase2 = train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, config.EPOCHS_PHASE2, DEVICE, config.EARLY_STOPPING_PATIENCE, config.CLIP_GRAD_NORM, model_name="best_model_late_fusion_eurosat_phase2.pth") # Modified model name

    # Plot loss and accuracy curves for phase 2
    plot_loss_curves(train_losses_phase2, val_losses_phase2, 2)
    plot_accuracy_curves(train_accs_phase2, val_accs_phase2, 2)

    # --- Évaluation finale sur l'ensemble de test ---
    logging.info("\nÉvaluation finale sur l'ensemble de test")
    #model.eval() # Already in evaluate_model

    evaluate_model(model, test_loader, train_loader, val_loader, DEVICE, config.NUM_CLASSES)

    # Plot attention weights
    plot_attention_weights(model, test_loader, DEVICE, num_samples=100) # Visualiser les poids d'attention
```
