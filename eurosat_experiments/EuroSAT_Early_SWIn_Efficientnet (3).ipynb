```python
import os
import torch
import numpy as np
import random # Import random for Python seed
from torch import nn, optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
from tqdm import tqdm
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import timm
from torch.nn.init import xavier_uniform_
from PIL import Image
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import torch.nn.functional as F # Import functional for interpolate

# Constants (same as before)
IMG_SIZE = 224
BATCH_SIZE = 16  # REDUCED BATCH SIZE to 16
NUM_CLASSES = 10
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
EPOCHS_PHASE_1 = 10  # Number of epochs for phase 1
EPOCHS_PHASE_2 = 1  # Number of epochs for phase 2
PREPROCESSED_DIR = "/content/drive/MyDrive/preprocessed_eurosat"
dataset_fraction=1.0
EPOCHS = EPOCHS_PHASE_1 # Initial EPOCHS value, will be updated dynamically

from google.colab import drive
drive.mount('/content/drive')


# Transformations (same as before)
train_transforms = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.RandomCrop(IMG_SIZE, padding=4),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

val_transforms = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

test_transforms = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Dataset Class (same as before)
class EuroSATDataset(Dataset):
    def __init__(self, images, labels, transforms):
        self.images = images
        self.labels = labels
        self.transforms = transforms

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = self.images[idx]
        if not isinstance(image, np.ndarray):
            raise TypeError(f"Expected numpy array, got {type(image)}")
        image = Image.fromarray(image)
        image = self.transforms(image)
        label = torch.tensor(self.labels[idx], dtype=torch.long)
        return image, label

# CBAM Module (Corrected implementation of _initialize_weights and forward) - REMOVED

# Early-Level Fusion Model WITHOUT CBAM
class EarlyFusionModel(nn.Module):
    def __init__(self, num_classes, embed_dim=256): # Removed transformer parameters, simplified
        super().__init__()

        # Swin Transformer for global context (using early feature layer)
        self.swin = timm.create_model("swin_tiny_patch4_window7_224", pretrained=True, features_only=True)

        # EfficientNet (using early feature layer)
        self.efficientnet = timm.create_model("efficientnet_b0", pretrained=True, features_only=True)

        swin_feature_dim_layer1 = self.swin.feature_info.channels()[0] # Using 1st feature layer (index 0)
        efficientnet_feature_dim_layer1 = self.efficientnet.feature_info.channels()[0] # Using 1st feature layer (index 0)

        # Fusion Convolutional Layers - Simple approach after concatenation
        fusion_input_dim = swin_feature_dim_layer1 + efficientnet_feature_dim_layer1 # Correct fusion input dimension for layer 1
        self.fusion_conv = nn.Sequential(
            nn.Conv2d(fusion_input_dim, 2 * embed_dim, kernel_size=3, padding=1), # Corrected input channels
            nn.BatchNorm2d(2 * embed_dim),
            nn.ReLU(inplace=True),
            nn.Conv2d(2 * embed_dim, embed_dim, kernel_size=3, padding=1),
            nn.BatchNorm2d(embed_dim),
            nn.ReLU(inplace=True)
        )


        # Adaptive Average Pooling
        self.avg_pool = nn.AdaptiveAvgPool2d((1, 1)) # Single pooling after fusion

        # Classification head
        self.classifier = nn.Linear(embed_dim, num_classes) # Input is now embed_dim
        self.classifier_dropout = nn.Dropout(0.3)

        self._init_weights()

    def _init_weights(self):
        for m in [self.classifier, self.fusion_conv[0], self.fusion_conv[3]]: # Initialize conv layers as well, removed proj layers
            if isinstance(m, nn.Linear):
                xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0.)
            elif isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0.)
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)


    def forward(self, image):
        # 1. Feature Extraction
        swin_features_list = self.swin(image)
        efficientnet_features_list = self.efficientnet(image)

        # Early features (using 1st layer - index 0) - CHANGED TO LAYER 1
        swin_features_early = swin_features_list[0] # Shape: [B, H, W, C] - Layer 1
        efficientnet_features_early = efficientnet_features_list[0] # Shape: [B, C, H, W] - Layer 1

        # Permute Swin to [B, C, H, W] to match EfficientNet and for concatenation
        swin_features_early = swin_features_early.permute(0, 3, 1, 2)

        # Resize feature maps to have the same spatial dimensions before concatenation.
        # Let's resize to the larger spatial size. We can pick the size of efficientnet features for simplicity
        swin_features_resized = F.interpolate(swin_features_early, size=efficientnet_features_early.shape[2:], mode='bilinear', align_corners=False)


        # 2. Early Fusion (Concatenation)
        fused_features_early = torch.cat([efficientnet_features_early, swin_features_resized], dim=1) # Concatenate along channel dimension
        #print("Shape after early fusion (concatenation):", fused_features_early.shape)

        # 3. Fusion Convolutional Layers
        fused_features_conv = self.fusion_conv(fused_features_early)
        #print("Shape after fusion conv layers:", fused_features_conv.shape)

        # 4. Pooling
        pooled_features = self.avg_pool(fused_features_conv)
        #print("Shape after pooling:", pooled_features.shape)


        flattened_features = pooled_features.squeeze(-1).squeeze(-1)
        #print("Shape of flattened_features:", flattened_features.shape)


        # 6. Classification
        projected_features_dropout = self.classifier_dropout(flattened_features) # Use flattened_features directly
        output = self.classifier(projected_features_dropout)

        return output


# Training, Validation and Evaluate functions (modified to NOT collect attention maps) - NO CHANGES NEEDED - REUSING FROM ORIGINAL CODE
def train_one_epoch(model, dataloader, optimizer, criterion, phase="Phase"):
    model.train()
    train_loss, correct, total = 0, 0, 0
    optimizer.zero_grad()
    for images, labels in tqdm(dataloader, desc=f"Training {phase}"): # Dynamic phase description
        images, labels = images.to(DEVICE), labels.to(DEVICE)

        outputs = model(images) # Get only output
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        train_loss += loss.item()
        _, predicted = outputs.max(1)
        correct += predicted.eq(labels).sum().item()
        total += labels.size(0)
    return train_loss / len(dataloader), 100. * correct / total


def validate(model, dataloader, criterion, phase="Phase"): # Dynamic phase description
    model.eval()
    val_loss, correct, total = 0, 0, 0
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for images, labels in tqdm(dataloader, desc=f"Validation {phase}"): # Dynamic phase description
            images, labels = images.to(DEVICE), labels.to(DEVICE)
            outputs = model(images) # Get only output
            loss = criterion(outputs, labels)
            val_loss += loss.item()
            _, predicted = outputs.max(1)
            correct += predicted.eq(labels).sum().item()
            total += labels.size(0)
            all_preds.append(predicted.cpu())
            all_labels.append(labels.cpu())
    all_preds = torch.cat(all_preds).numpy()
    all_labels = torch.cat(all_labels).numpy()
    return val_loss / len(dataloader), 100. * correct / total, all_labels, all_preds


def evaluate_model(model, dataloader, num_classes, set_name="Validation Phase"): # Dynamic set_name
    """Evaluates the model."""
    model.eval()
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for images, labels in tqdm(dataloader, desc=f"Evaluating on {set_name}"): # Dynamic set_name
            images, labels = images.to(DEVICE), labels.to(DEVICE)
            outputs = model(images) # Get only output
            _, predicted = torch.max(outputs, 1)
            all_preds.append(predicted.cpu())
            all_labels.append(labels.cpu())
    all_preds = torch.cat(all_preds).numpy()
    all_labels = torch.cat(all_labels).numpy()

    print(f"\n{set_name} Classification Report:")
    report = classification_report(all_labels, all_preds, target_names=[str(i) for i in range(num_classes)], digits=4)
    print(report)

    cm = confusion_matrix(all_labels, all_preds, normalize='true')
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt=".2%", cmap="Blues")
    plt.xlabel("Predicted Labels", fontsize=12, color='black') # English text
    plt.ylabel("True Labels", fontsize=12, color='black')    # English text
    plt.title(f"{set_name} Confusion Matrix (Percentage)", fontsize=14, color='black') # English text
    plt.show()
    return all_labels, all_preds


# Load preprocessed data function (no changes needed) - REUSING FROM ORIGINAL CODE
def load_or_preprocess_data(dataset_fraction=1.0, train_split=0.8, val_split=0.1, test_split=0.1):
    """Load or preprocess the EuroSAT dataset."""
    if train_split + val_split + test_split != 1.0:
        raise ValueError("Train, validation, and test splits must sum to 1.0")
    try:
        if os.path.exists(os.path.join(PREPROCESSED_DIR, "train_data.npz")):
            print("Loading preprocessed data...")
            with np.load(os.path.join(PREPROCESSED_DIR, "train_data.npz")) as data:
                train_images, train_labels = data['images'], data['labels']
            with np.load(os.path.join(PREPROCESSED_DIR, "val_data.npz")) as data:
                val_images, val_labels = data['images'], data['labels']
            with np.load(os.path.join(PREPROCESSED_DIR, "test_data.npz")) as data:
              test_images, test_labels = data['images'], data['test_labels']
        else:
            print("Processing dataset...")
            total_samples = 27000
            samples_to_load = int(total_samples * dataset_fraction)
            dataset = tfds.load("eurosat", split=f"train[:{samples_to_load}]", as_supervised=True)
            dataset = list(tfds.as_numpy(dataset))
            train_end = int(train_split * samples_to_load)
            val_end = int((train_split + val_split) * samples_to_load)
            train_data = dataset[:train_end]
            val_data = dataset[train_end:val_end]
            test_data = dataset[val_end:]
            def process_dataset(data):
                images, labels = [], []
                for img, label in data:
                    images.append(img)
                    labels.append(label)
                return np.stack(images), np.array(labels)
            train_images, train_labels = process_dataset(train_data)
            val_images, val_labels = process_dataset(val_data)
            test_images, test_labels = process_dataset(test_data)
            os.makedirs(PREPROCESSED_DIR, exist_ok=True)
            np.savez_compressed(os.path.join(PREPROCESSED_DIR, "train_data.npz"), images=train_images, labels=train_labels)
            np.savez_compressed(os.path.join(PREPROCESSED_DIR, "val_data.npz"), images=val_images, labels=val_labels)
            np.savez_compressed(os.path.join(PREPROCESSED_DIR, "test_data.npz"), images=test_images, test_labels=test_labels)
    except FileNotFoundError as e:
        print(f"File not found error: {e}")
        raise e
    except Exception as e:
        print(f"Unexpected error: {e}")
        raise e
    return train_images, train_labels, val_images, val_labels, test_images, test_labels


if __name__ == "__main__":
    torch.cuda.empty_cache() # Clear CUDA cache at the start
    phase_to_run = 0 # Set to 0 to run both phases if no pre-trained model is found, or 1 or 2 to run specific phase

    # Set seeds for reproducibility
    torch.manual_seed(42)
    np.random.seed(42)
    random.seed(42)
    torch.cuda.manual_seed_all(42)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

    # Model save paths
    model_save_path_phase1 = os.path.join("/content/drive/MyDrive", "best_early_fusion_eurosat_phase1_latest.pth") # Changed model name in path
    model_save_path_phase2 = os.path.join("/content/drive/MyDrive", "best_early_fusion_eurosat_phase2_latest.pth") # Changed model name in path
    model_save_path = model_save_path_phase2 if phase_to_run == 2 else model_save_path_phase1 # Dynamic save path

    model = EarlyFusionModel(NUM_CLASSES).to(DEVICE) # Instantiate EarlyFusionModel

    # # Debug: Print feature info
    # print("Swin Feature Info:", model.swin.feature_info)
    # print("EfficientNet Feature Info:", model.efficientnet.feature_info)
    # print("Swin Feature Channels:", model.swin.feature_info.channels())
    # print("Swin Feature Reductions:", model.swin.feature_info.reduction())
    # print("EfficientNet Feature Channels:", model.efficientnet.feature_info.channels())
    # print("EfficientNet Feature Reductions:", model.efficientnet.feature_info.reduction())


    model_loaded = False # Flag to track if model is loaded

    # Check for pre-trained models and load if available, set model_loaded flag
    if os.path.exists(model_save_path_phase2):
        print(f"Loading pre-trained Phase 2 model from: {model_save_path_phase2}")
        model.load_state_dict(torch.load(model_save_path_phase2))
        print("Pre-trained Phase 2 model loaded successfully. Skipping training phases.")
        model_loaded = True
    elif os.path.exists(model_save_path_phase1):
        print(f"Loading pre-trained Phase 1 model from: {model_save_path_phase1}")
        model.load_state_dict(torch.load(model_save_path_phase1))
        print("Pre-trained Phase 1 model loaded successfully. Skipping Phase 1 training.")
        model_loaded = True # Still mark as loaded, even if Phase 1 only


    # Load data and dataloaders (common for both cases)
    train_images, train_labels, val_images, val_labels, test_images, test_labels = load_or_preprocess_data(dataset_fraction)
    train_dataset = EuroSATDataset(train_images, train_labels, train_transforms)
    val_dataset = EuroSATDataset(val_images, val_labels, val_transforms)
    test_dataset = EuroSATDataset(test_images, test_labels, test_transforms)
    train_dataloader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, pin_memory=True, num_workers=2)
    val_dataloader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, pin_memory=True, num_workers=2)
    test_dataloader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, pin_memory=True, num_workers=2)


    if not model_loaded or phase_to_run == 1: # Run Phase 1 if no model loaded OR if phase_to_run is set to 1
        EPOCHS = EPOCHS_PHASE_1 # Set EPOCHS for phase 1
        if not model_loaded:
            print("No pre-trained model found. Starting Phase 1 training from scratch.")
            # Freeze lower layers for Phase 1 (only if starting from scratch) - ADJUSTED FOR EARLY FUSION MODEL
            for name, param in model.named_parameters():
                if "swin.patch_embed" in name or "swin.layers" in name[:1] : # Freezing very early swin layers - reduced to first layer block only
                    param.requires_grad = False
                if "efficientnet.conv_stem" in name or "efficientnet.blocks" in name[:2]: # Freezing very early efficientnet layers - reduced to first few blocks
                    param.requires_grad = False
                if "fusion_conv" in name or "classifier" in name: # Train fusion and classifier
                    param.requires_grad = True
        else:
            print("Continuing Phase 1 training from loaded model.") # Model loaded, but running Phase 1 anyway (if phase_to_run==1)

        # Optimizer for Phase 1: Layer-specific learning rates - ADJUSTED FOR EARLY FUSION MODEL
        optimizer_phase1 = optim.Adam([
             {'params': model.efficientnet.parameters(), 'lr': 1e-5}, # LR plus faible pour EfficientNet backbone
             {'params': model.swin.parameters(), 'lr': 1e-5},        # LR plus faible pour Swin backbone
             {'params': model.fusion_conv.parameters(), 'lr': 1e-4}, # LR for fusion conv layers
             {'params': model.classifier.parameters(), 'lr': 1e-4},    # LR raisonnable pour classifier
         ])
        scheduler_phase1 = optim.lr_scheduler.ReduceLROnPlateau(optimizer_phase1, mode='min', patience=3, factor=0.5, verbose=True)
        criterion = nn.CrossEntropyLoss()

        best_val_loss_phase1 = float('inf')
        patience_counter_phase1 = 0
        early_stopping_patience = 5
        train_losses_phase1, val_losses_phase1, train_accuracies_phase1, val_accuracies_phase1 = [], [], [], []

        for epoch in range(EPOCHS):
            train_loss, train_acc = train_one_epoch(model, train_dataloader, optimizer_phase1, criterion, phase="Phase 1")
            val_loss, val_acc, _, _ = validate(model, val_dataloader, criterion, phase="Phase 1")
            train_losses_phase1.append(train_loss)
            val_losses_phase1.append(val_loss)
            train_accuracies_phase1.append(train_acc)
            val_accuracies_phase1.append(val_acc)

            print(f"Epoch {epoch + 1} (Phase 1): Train Loss={train_loss:.4f}, Train Acc={train_acc:.4f}, Val Loss={val_loss:.4f}, Val Acc={val_acc:.4f}")
            scheduler_phase1.step(val_loss)

            if val_loss < best_val_loss_phase1:
                best_val_loss_phase1 = val_loss
                patience_counter_phase1 = 0
                torch.save(model.state_dict(), model_save_path_phase1)
                print(f"Best Phase 1 model saved to: {model_save_path_phase1}")
            else:
                patience_counter_phase1 += 1
                print(f"Early stopping patience counter (Phase 1): {patience_counter_phase1}/{early_stopping_patience}")

            if patience_counter_phase1 >= early_stopping_patience:
                print("Early stopping triggered for Phase 1.")
                break

        model.load_state_dict(torch.load(model_save_path_phase1)) # Load best phase 1 model after training
        val_loss_phase1_final, val_acc_phase1_final, val_labels_phase1_final, val_preds_phase1_final = validate(model, val_dataloader, criterion, phase="Phase 1")
        print(f"Best Validation Loss (Phase 1): {best_val_loss_phase1:.4f}, Final Validation Acc (Phase 1): {val_acc_phase1_final:.4f}")
        evaluate_model(model, val_dataloader, NUM_CLASSES, set_name="Validation Phase 1")
        evaluate_model(model, test_dataloader, NUM_CLASSES, set_name="Test Phase 1")


    if not model_loaded or phase_to_run == 0 or phase_to_run == 2: # Run Phase 2 if no model loaded OR phase_to_run is 0 (both) or 2
        EPOCHS = EPOCHS_PHASE_2 # Update EPOCHS for phase 2
        if not model_loaded:
            print("Starting Phase 2 Training after Phase 1.") # Indicate phase 2 start after phase 1
            # Unfreeze more layers for Phase 2 (if coming from Phase 1 training) - ADJUSTED FOR EARLY FUSION MODEL
            for name, param in model.named_parameters():
                param.requires_grad = True # Unfreeze all parameters for Phase 2 in early fusion model
        else:
             print("Continuing Phase 2 training from loaded model.") # Model loaded, continuing phase 2 if phase_to_run == 2

        # Optimizer for Phase 2: Layer-specific learning rates (smaller than Phase 1) - ADJUSTED FOR EARLY FUSION MODEL
        optimizer_phase2 = optim.Adam([
             {'params': model.efficientnet.parameters(), 'lr': 5e-6}, # LR encore plus faible pour EfficientNet backbone
             {'params': model.swin.parameters(), 'lr': 5e-6},        # LR encore plus faible pour Swin backbone
             {'params': model.fusion_conv.parameters(), 'lr': 5e-5}, # LR for fusion conv layers
             {'params': model.classifier.parameters(), 'lr': 5e-5},    # LR raisonnable pour classifier
         ])
        scheduler_phase2 = optim.lr_scheduler.ReduceLROnPlateau(optimizer_phase2, mode='min', patience=3, factor=0.5, verbose=True)
        criterion = nn.CrossEntropyLoss()

        best_val_loss_phase2 = float('inf')
        patience_counter_phase2 = 0
        early_stopping_patience_phase2 = 5
        train_losses_phase2, val_losses_phase2, train_accuracies_phase2, val_accuracies_phase2 = [], [], [], []

        for epoch in range(EPOCHS):
            train_loss, train_acc = train_one_epoch(model, train_dataloader, optimizer_phase2, criterion, phase="Phase 2")
            val_loss, val_acc, _, _ = validate(model, val_dataloader, criterion, phase="Phase 2")
            train_losses_phase2.append(train_loss)
            val_losses_phase2.append(val_loss)
            train_accuracies_phase2.append(train_acc)
            val_accuracies_phase2.append(val_acc)

            print(f"Epoch {epoch + 1} (Phase 2): Train Loss={train_loss:.4f}, Train Acc={train_acc:.4f}, Val Loss={val_loss:.4f}, Val Acc={val_acc:.4f}")
            scheduler_phase2.step(val_loss)

            if val_loss < best_val_loss_phase2:
                best_val_loss_phase2 = val_loss
                patience_counter_phase2 = 0
                torch.save(model.state_dict(), model_save_path_phase2)
                print(f"Best Phase 2 model saved to: {model_save_path_phase2}")
            else:
                patience_counter_phase2 += 1
                print(f"Early stopping patience counter (Phase 2): {patience_counter_phase2}/{early_stopping_patience_phase2}")

            if patience_counter_phase2 >= early_stopping_patience_phase2:
                print("Early stopping triggered for Phase 2.")
                break

        model.load_state_dict(torch.load(model_save_path_phase2)) # Load best phase 2 model
        val_loss_phase2_final, val_acc_phase2_final, val_labels_phase2_final, val_preds_phase2_final = validate(model, val_dataloader, criterion, phase="Phase 2")
        print(f"Best Validation Loss (Phase 2): {best_val_loss_phase2:.4f}, Final Validation Acc (Phase 2): {val_acc_phase2_final:.4f}")

        test_labels_final, test_preds_final = evaluate_model(model, test_dataloader, NUM_CLASSES, set_name="Test Phase 2")
        evaluate_model(model, val_dataloader, NUM_CLASSES, set_name="Validation Phase 2")


    print("\nTraining Complete.")
    if not model_loaded or phase_to_run == 0: # Print message if both phases ran or started from scratch
        print(f"Best model for Phase 1 saved as: {model_save_path_phase1}")
        print(f"Best model for Phase 2 saved as: {model_save_path_phase2}")
    elif phase_to_run == 1:
        print(f"Best model for Phase 1 saved as: {model_save_path_phase1} (Phase 2 skipped)")
    elif phase_to_run == 2:
         print(f"Best model for Phase 2 saved as: {model_save_path_phase2} (Phase 1 skipped or loaded)")

    print("Training and evaluation completed without attention map visualizations.")
```
